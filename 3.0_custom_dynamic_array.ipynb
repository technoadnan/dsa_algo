```python
import ctypes
class ad_list:
   def __init__(self) -> None:
      self.size = 1 # max input it can take 
      self.n = 0  # input those are in size
      self.A = self.__make_array(self.size)
   
   def __len__(self):
      return self.n

   def __print__(self):
      return self.A
   
   def append(self, item):
      if self.n == self.size:
         self.__resize(self.size * 2)
      self.A[self.n] = item
      self.n = self.n + 1

   def __resize(self, new_capacity):
      # new array
      B = self.__make_array(new_capacity) # skeleton 
      self.size = new_capacity
      # copy
      for i in range(self.n):
         B[i] = self.A[i]
      # re-assgin 
      self.A = B


   def __str__(self) -> str:
      result = ''
      for j in range(self.n):
         result = result + str(self.A[j]) + ","
      return f"[{result[:-1]}]"
   
   def __getitem__(self, index):
      if 0 <= index < self.n:
         return self.A[index]
      else:
         raise ValueError
   
   def pop(self):
      if self.n == 0:
         raise IndexError
      self.n = self.n - 1 
      return self.A[self.n]

   def clear(self):
      self.n = 0
      return "[]"

   def finds(self, item):
      for j in range(self.n):
         if item == self.A[j]:
            return j
         else:
            return "not in list"
   
   def insert(self, pos, item):
      

      if self.n == self.size:
         self.__resize(self.size * 2)

      for i in range(self.n, pos, -1):
         self.A[i] = self.A[i - 1]
      
      self.A[pos] = item
      self.n = self.n + 1

   def __delitem__(self, pos):
      
      if 0 <= pos <= self.n-1: # otherwise n - 1 will work despit the pos is not in range
         for i in range(pos, self.n-1):
            self.A[i] = self.A[i + 1]

         self.n = self.n - 1

   def remove(self, item):
      pos = self.finds(item=item)
      if type(pos) == int:
         self.__delitem__(pos=pos)
      else:
         return pos 

   def __make_array(self, capacity):
      return (capacity*ctypes.py_object)()
   
```
